```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plotting test accuracy vs. pruning fraction 

stats = pd.read_csv("quantized_readings/models_stats_quantized.csv")
pruning_fractions = stats['Pruning Fraction']
test_accuracies = stats['Test Accuracy']

fig = plt.figure(figsize=(10, 5))
plt.plot(pruning_fractions, test_accuracies)
plt.xlabel('Pruning Fraction')
plt.ylabel('Test Accuracy (%)')
plt.title('MobileNetV1: Test Accuracy vs. Pruning Fraction (Quantized)')
plt.grid()
plt.show()
```

```python
# plotting test accuracy vs. number of parameters 

num_params = stats['# Params']

fig = plt.figure(figsize=(10, 5))
plt.plot(num_params, test_accuracies)
plt.xlabel('Number of Parameters')
plt.ylabel('Test Accuracy (%)')
plt.title('MobileNetV1: Test Accuracy vs. Number of Parameters (Quantized)')
plt.grid()
plt.show()
```

```python
#  plotting total inference time vs. pruning fraction 

stats = pd.read_csv("quantized_readings/models_stats_quantized.csv")
pruning_fractions = stats['Pruning Fraction']
inference_times = stats['Inference Time']

fig = plt.figure(figsize=(10, 5))
plt.plot(pruning_fractions, inference_times)
plt.xlabel('Pruning Fraction')
plt.ylabel('Inference Time [s]')
plt.title('MobileNetV1: Inference Time vs. Pruning Fraction (Quantized)')
plt.grid()
plt.show()
```
