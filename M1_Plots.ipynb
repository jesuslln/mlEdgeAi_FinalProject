```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plotting test accuracy vs. pruning fraction 

stats = pd.read_csv("onnx_rpi_readings/models_stats.csv")
pruning_fractions = stats['Pruning Fraction']
test_accuracies = stats['Test Accuracy']

fig = plt.figure(figsize=(10, 5))
plt.plot(pruning_fractions, test_accuracies)
plt.xlabel('Pruning Fraction')
plt.ylabel('Test Accuracy (%)')
plt.title('MobileNetV1: Test Accuracy vs. Pruning Fraction')
plt.grid()
plt.show()
```

```python
# plotting test accuracy vs. number of parameters 

num_params = stats['# Params']

fig = plt.figure(figsize=(10, 5))
plt.plot(num_params, test_accuracies)
plt.xlabel('Number of Parameters')
plt.ylabel('Test Accuracy (%)')
plt.title('MobileNetV1: Test Accuracy vs. Number of Parameters')
plt.grid()
plt.show()
```
